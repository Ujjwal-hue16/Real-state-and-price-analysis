# **Analysis and Visualization**

```python
df.head()
```

```python
df.id.value_counts().count() 
```

```python
rooms_lot= df.groupby('bedrooms')['sqft_lot'].count().reset_index()
rooms_lot
```

```python
px.scatter(rooms_lot ,x='sqft_lot',y='bedrooms',title='Relationship between Sqft Lot and Bedrooms')
```

```python
bedroom_price = df.groupby('bedrooms')['price'].mean().reset_index()
bedroom_price 
```

```python
px.line(bedroom_price ,x='bedrooms',y='price',title='Price trends to no of Bedrooms')
```

```python
px.scatter(df,'price') 
```

```python
date_price=df.groupby('date')['price'].count().reset_index()
date_price
```

```python
px.line(date_price,x='date',y='price',title ='Data vs Price')
```

```python
bathroom_bedroom = df.groupby('bathrooms')['bedrooms'].mean().reset_index().astype(int)
bathroom_bedroom
```

```python
px.bar( bathroom_bedroom,x='bedrooms',y='bathrooms',title='Bedrooms vs Bathrooms') 
```

```python
bathroom_price = df.groupby('bathrooms')['price'].mean().reset_index().astype(int)
bathroom_price 
```

```python
px.bar( bathroom_price,x='bathrooms',y='price',title='Price vs Bathrooms')  
```

```python
floors_price= df.groupby('floors')['price'].mean().reset_index().astype(int)
floors_price
```

```python
px.line( floors_price,x='floors',y='price',title='Price trends to no of Floors')
```

```python
waterfront_view = pd.crosstab(df['waterfront'], df['view'])
waterfront_view 
```

```python
px.imshow(waterfront_view,labels=dict(x="View", y="Waterfront", color="Count"),
                x=waterfront_view.columns,
                y=waterfront_view.index,
                color_continuous_scale='viridis') 
```

```python
grade_condition =pd.crosstab(df['grade'], df['condition'])
grade_condition 
```

```python
px.imshow(grade_condition,
                 labels=dict(x="Grade", y="Condition", color="Count"),
                 x=grade_condition.columns,
                 y=grade_condition.index,
                 color_continuous_scale='viridis',
                 title='Cross-tabulation between Grade and Condition') 
```

```python
zipcode_price = df.groupby('zipcode')['price'].median().reset_index()
zipcode_price
```

```python
px.bar(zipcode_price, x='zipcode', y='price', title='House Price by Zipcode')  
```

```python
yr_built_price = df.groupby('yr_built')['price'].median().reset_index()
yr_built_price 
```
